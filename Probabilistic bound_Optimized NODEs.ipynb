```python
# Re-importing the necessary libraries
import matplotlib.pyplot as plt
import math
from scipy.special import comb, factorial
import numpy as np

# Global figure object
fig, ax = plt.subplots(figsize=(10, 6))

def compute_upper_bound(N, k, d):
    """
    Compute the upper bound for the probability P(M <= k) with checks for valid arguments.

    :param N: Natural number N
    :param k: Value k (1 to 2N-1)
    :param d: Exponent d
    :return: Upper bound for the probability
    """
    # Calculating the summations with checks for valid binomial coefficient arguments
    first_sum = sum(comb(N - 1, p - 1)**2 for p in range(math.ceil(k / 2 + 1), N))
    second_sum = sum(comb(N - 1, p) * comb(N - 1, p - 1) for p in range(math.ceil((k + 1)/ 2), N - 1))

    # Total expression inside the bracket
    expression_inside_bracket = first_sum + second_sum

    # The fraction part
    fraction_part = (2 * factorial(N)**2) / factorial(2 * N)

    # The complete expression
    complete_expression = (expression_inside_bracket * fraction_part)**d

    # Calculate the upper bound
    upper_bound = 1 - complete_expression

    return upper_bound

def plot_upper_bounds(N, d_values):
    global fig, ax
    cmap = plt.cm.get_cmap('viridis', len(d_values))

    for i, d in enumerate(d_values):
        k_values = range(1, 2 * N - 1)
        upper_bounds = [compute_upper_bound(N, k, d) for k in k_values]
        modk_values = [k - 1 for k in k_values]

        color = cmap(i)
        ax.plot(modk_values, upper_bounds, marker='o', color=color, label=f"d={d}")

    ax.set_xlabel(r'$0\leq k\leq 2N-2$', fontsize=14)
    ax.set_ylabel("Lower Bound", fontsize=14)
    ax.set_yticks(np.arange(0, 1.1, 0.1))
    ax.set_xticks([x for x in range(0, 2 * N - 1)])
    ax.grid(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Additional legend for N
    ax.text(0.85, 0.1, f'$N={N}$', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

    # Positioning the legend above the ax.text()
    ax.legend(loc='lower center', bbox_to_anchor=(0.85, 0.15),ncol=1)

def save_figure_as_pdf(filename):
    global fig
    fig.savefig(filename, format='pdf')

# Example usage
d_values = [2,8, 20, 100, 2000]
plot_upper_bounds(10, d_values)  
save_figure_as_pdf("lowerbound1.pdf")  # Save as PDF

plt.show()  # Display the plot

```
